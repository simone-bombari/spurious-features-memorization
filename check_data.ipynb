## A Figure for the paper

```python
data_name = 'MNIST'
train_dir = './data/' + os.path.join(data_name, 'train')
augmented_data_train = np.load(train_dir + '/augmented_data.npy')
spurious_data = np.load(train_dir + '/spurious_data.npy')

fig, ax = plt.subplots(2, 2, figsize=(8, 8))

cat = mpimg.imread('cat.png')
cat_corrupted = mpimg.imread('cat_masked.png')

ax[0][0].imshow(cat)
ax[0][1].imshow(cat_corrupted)
ax[1][0].imshow(augmented_data_train[0])
ax[1][1].imshow(spurious_data[0])

for i in range(2):
    for j in range(2):
        ax[i][j].yaxis.set_ticks([])
        ax[i][j].yaxis.set_ticklabels([])
        ax[i][j].xaxis.set_ticks([])
        ax[i][j].xaxis.set_ticklabels([])

fig.tight_layout()
fig.savefig('example.pdf')
```
